### Prompt 4:

The data in the "df" DataFrame has been cleaned and processed. Now I want to start building new variables/columns that will provide good support for analyzing the company's sales performance.

The columns available in "df" are as follows:
- ItemID: A unique identifier for each item of clothing.
- ItemName: The name of the clothing item (e.g. T-shirt, jeans, jacket).
- Brand: The brand associated with the item of clothing (e.g. BrandA, BrandB).
- Category: The category under which the item of clothing falls (e.g. Topwear, Bottomwear, Outerwear).
- Size: The size of the item of clothing (e.g. XS, S, M, L, XL).
- Price: The price of the item of clothing in USD.

The aim is to create variables or columns that provide relevant information for the company's managers.

Based on the information provided, what types of variables/new columns would be useful to achieve the objective of analyzing sales performance?

```python
## PriceRange
# Classifies items into price ranges, which can be useful for market analysis and segmentation

df['PriceRange'] = pd.cut(df['Price'], bins=[0, 20, 50, 100, 200, 500], labels=['0-20', '21-50', '51-100', '101-200', '201-500'])
```

```python
## BrandCategory
# Combines brand and category to identify specific combinations

df['BrandCategory'] = df['Brand'] + ' - ' + df['Category']
```

```python
## SizeAvailability
# Creates a column indicating the availability of popular sizes (optional, depending on other available data)

popular_sizes = ['M', 'L', 'XL']
df['SizeAvailability'] = df['Size'].apply(lambda x: 'Popular' if x in popular_sizes else 'Non-Popular')
```

```python
## PricePerUnitSize
# Calculates the price per unit size, useful for comparing prices of items of different sizes

size_map = {'XS': 1, 'S': 2, 'M': 3, 'L': 4, 'XL': 5}  # Exemplo de mapeamento de tamanhos
df['PricePerUnitSize'] = df['Price'] / df['Size'].map(size_map)
```

```python
## CategorySalesWeight
# Assigns a weight to categories based on a specific criterion, such as the number of sales (if this data is available)

category_weights = {
    'Outerwear': 1.5,
    'Topwear': 1.2,
    'Bottomwear': 1.0,
    'Dresses': 1.3,
    'Footwear': 1.4
}
df['CategorySalesWeight'] = df['Category'].map(category_weights)
```

```python
## AveragePriceByBrand
# Calculates the average price of items by brand, useful for comparing the average price between different brands

avg_price_by_brand = df.groupby('Brand')['Price'].transform('mean')
df['AveragePriceByBrand'] = avg_price_by_brand
```

```python
## IsPremiumBrand
# Identifies whether the brand is considered premium (based on a specific criterion such as average price above a certain value)

premium_threshold = 100  # Example of a reference value for premium brands
df['IsPremiumBrand'] = df['AveragePriceByBrand'] > premium_threshold
```

```python
## PriceVarianceByCategory
# Calculates the variance of prices within each category, useful for understanding price dispersion

price_variance_by_category = df.groupby('Category')['Price'].transform('var')
df['PriceVarianceByCategory'] = price_variance_by_category
```

### Prompt 6:

I want to create a visualization that allows me to identify the total obtained in sales by product categories in a horizontal bar chart.

The data needs to be grouped by category.

The figure in the plot must follow the store's visual identity, with the background of the figure and the plot in the color "#ebebeb".

The bars must be colored "#e23155".

The bar values need to be ordered to maintain a funnel look.

In addition, you need to display the total sales in millions inside or next to each bar, with the text of this information in the color "#19325c" and in the structure "R$ value M".

Don't show the grid in the plot.

Use Python's data visualization libraries to create this visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Group data by category and calculate total sales by category
total_sales_by_category = df.groupby('Category')['Price'].sum().reset_index()

# Sort the categories by total sales in descending order
total_sales_by_category = total_sales_by_category.sort_values(by='Price', ascending=False)

# Style setting
sns.set(style="whitegrid")
plt.rcParams['figure.facecolor'] = '#ebebeb'
plt.rcParams['axes.facecolor'] = '#ebebeb'

# Adjust the size of the picture
plt.figure(figsize=(14, 6))  # I increased the width and decreased the height for thinner bars

# Create the horizontal bar chart
bars = plt.barh(total_sales_by_category['Category'], total_sales_by_category['Price'], color='#e23155', height=0.5)

# Add the sales figures in millions next to each bar
for bar in bars:
    width = bar.get_width()
    label = f"R$ {width / 1e6:.2f} M"
    plt.text(width, bar.get_y() + bar.get_height() / 2, label, va='center', ha='left', color='#19325c', fontsize=12)

# Adjust title and labels
plt.title('Total de Vendas por Categoria de Produtos', fontsize=16)
plt.xlabel('Total de Vendas (USD)', fontsize=14)
plt.ylabel('Categoria', fontsize=14)
plt.gca().invert_yaxis()  # Invert the y-axis to sort from top to bottom

# Remove the grid
plt.grid(False)

# Adjust spacing
plt.tight_layout()

# Display the graphic
plt.show()
```
